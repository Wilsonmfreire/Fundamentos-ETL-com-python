```python
import pandas as pd
```

```python
valores_ausentes = ['***', '****' , '*****', '####','NULL']
df = pd.read_csv("ocorrencia.csv", sep = "," , parse_dates=["ocorrencia_dia"], dayfirst= True, na_values= valores_ausentes)
df.head(5)
```

Comando loc-
Para ir até uma coluna ou linha como o select
Dentro dos [ ] passamos dois parametros, a primeira seria LInha e a segunda nome da coluna. Usando a tabela a cima vamos ter "Belém" 
                                         passando como parametro ex [1:2] vamos ter da linha 1 a linha 2, se passar um valor maior como segundo ira aumentar a entrega [1:20]
                                         passando como parametro [[10,40]] como resposta duas linhas expecificas da tabela  a linha 10 e a 40.
                                         passando como parametro [:,'ocorrencia_cidade'] vamos ter a todas as linhas da coluna informada

```python
df.loc[1,'ocorrencia_cidade']
```

```python
df.loc[1:2]
```

```python
df.loc[[10,40]]
```

```python
df.loc[:,'ocorrencia_cidade']
```

Para limpar uma lista de informações-
Usando o comando .replace(['', '', '',''], pd.NA, inplace= TRUE)

Lembrando que pd.NA do Pandas para informar campos nulls e que essa transformação fique registrada inplace = True

```python
df.replace(['***', '****' , '*****', '####','NULL'], pd.NA, inplace=True)
```

Comando df.dropna() sem passar nada como parametro ele ira buscar na tabela toda os valores NA e excluir
Comando df.dropna(subset=['ocorrencia_aerodromo']) dessa maneira ele ira procurar somente na coluna indicada.
obs: lembrando que as alterações só serão feitas com o comando inplace=True

```python
df.dropna()
```
